```python
%matplotlib inline
%pylab inline
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
```

```python
df_pagerank = pd.read_csv('pagerank.csv', sep=',')
df_in = pd.read_csv('in.csv', sep=',')
df_out = pd.read_csv('out.csv', sep=',')
df_wcc = pd.read_csv('wcc.csv', sep=',')
df_kcore = pd.read_csv('kcore.csv', sep=',')

del df_pagerank['delta']

df_pagerank.columns = ['address', 'pagerank']
df_in.columns = ['address', 'in_degree']
df_out.columns = ['address', 'out_degree']
df_wcc.columns = ['address', 'component_id']
df_kcore.columns = ['address', 'core_id']
```

```python
print(len(df_in))
print(len(df_out))
print(len(df_pagerank))
```

```python
in_freq = df_in.groupby('in_degree', as_index=False).count()
out_freq = df_out.groupby('out_degree', as_index=False).count()
pagerank_freq = df_pagerank.groupby('pagerank', as_index=False).count()
wcc_freq = df_wcc.groupby('component_id', as_index=False).count()
kcore_freq = df_kcore.groupby('core_id', as_index=False).count()
```

```python
print(wcc_freq.head().sort_values('address', ascending=False))
```

```python
plt.hist(wcc_freq['address'],wcc_freq['component_id'])
plt.title('component_id')
plt.show()
```

```python
plt.scatter(in_freq['address'],in_freq['in_degree'],s=40, facecolor='r', lw = 0)
plt.title('in_degree')
plt.show()
```

```python
plt.scatter(out_freq['address'],out_freq['out_degree'],s=40, facecolor='b', lw = 0)
plt.title('out_degree')
plt.show()
```

```python
plt.scatter(pagerank_freq['address'],pagerank_freq['pagerank'],s=40, facecolor='g', lw = 0)
plt.title('pagerank')
plt.show()
```
